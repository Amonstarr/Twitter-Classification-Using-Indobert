```python
import pandas as pd
from sklearn.utils import shuffle

def balancing_df(path=None):
  if path == None:
    raise "No data"

  df = pd.read_csv(path)
  minimum = [len(df[df["label"] == 2]), len(df[df["label"] == 1]), len(df[df["label"] == 0])]
  minimum = min(minimum)
  df_0 = df[df["label"] == 0].sample(minimum)
  df_1 = df[df["label"] == 1].sample(minimum)
  df_2 = df[df["label"] == 2].sample(minimum)

  df_gabung = pd.concat([df_0, df_1, df_2])
  df_gabung = shuffle(df_gabung)
  df_gabung = df_gabung.reset_index(drop=True)

  return df_gabung
```

```python
import pandas as pd
from sklearn.utils import resample

def oversample_light(path=None, target_min=150):
    df = pd.read_csv(path)

    dfs = []

    for label, group in df.groupby("label"):
        if len(group) < target_min:
            group_resampled = resample(
                group,
                replace=True,
                n_samples=target_min,
                random_state=42
            )
            dfs.append(group_resampled)
        else:
            dfs.append(group)

    df_balanced = pd.concat(dfs)
    df_balanced = df_balanced.sample(frac=1, random_state=42).reset_index(drop=True)

    return df_balanced
```

```python
import os

balanced_df = oversample_light(
    "/content/drive/MyDrive/Satria Data/Unbalance/Test 1/dataset_penyisihan_bdc_2024_clean.csv",
    target_min=150
)

os.makedirs(OUTPUT_BASE, exist_ok=True)
balanced_df.to_csv(
    "/content/drive/MyDrive/Satria Data/Balance/clean_train_balanced.csv",
    index=False
)
```
